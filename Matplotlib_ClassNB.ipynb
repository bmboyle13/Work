## 1 Histogram : We should use histogram when we need the count of the variable in a plot.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df_hist=pd.read_csv('national_parks.csv')
```

```python
df_hist.head()
```

```python
plt.hist(df_hist['GrandCanyon'],
        facecolor='black',
        edgecolor='lightgreen',
        bins=10)
plt.title('Histogram plot of Grand Canyon')
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.show()
```

### Components of a histogram
* <b>n:</b> Contains the frequency of each bin
* <b>bins:</b> Represents the middle value of each bin
* <b>patches:</b> The Patch object for the rectangle shape representing each bar

```python
n,bins,patches=plt.hist(df_hist['GrandCanyon'],
        facecolor='lightgreen',
        edgecolor='blue',
        bins=10)
print('n',n)
print('bins',bins)
print('patches',patches)
plt.title('Histogram plot of Grand Canyon')
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.show()
```

#### The cumulative property
If True, then a histogram is computed where each bin gives the counts in that bin plus all bins for smaller values. The last bin gives the total number of datapoints.

```python
df_hist.columns
```

```python
plt.hist(df_hist['BryceCanyon'],
        facecolor='indigo',
        edgecolor='yellow',
        bins=10,
        cumulative=True)
plt.title('Histogram plot of Bryce Canyon')
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.show()
```

#### Check the histogram to a range of values
We only look at the data points within the range 2M-5M. This realigns the bins in the histogram

```python
df_hist.describe()
```

```python
plt.hist(df_hist['GrandCanyon'],
        facecolor='lightgreen',
        edgecolor='blue',
        bins=10,
        range=(2000000,5000000))
plt.title('Histogram plot of Grand Canyon range of 2M-5M')
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.show()
```

## 2  Pie Chart: Pie chart shows distribution of data based on proportion of pie occupied. 

#### Importing Libraries and Data

```python
import pandas as pd
import matplotlib.pyplot as plt
df_pie=pd.read_csv('types_movies.csv')
```

```python
df_pie.head()
```

```python
# use value_counts to count categorical values to plot pie chart
```

```python
plt.pie(df_pie['Percentage'],
       labels=df_pie['Sector'])
plt.title('Pie Chart of Movies Dataset')
plt.axis('equal')
plt.show()
```

### Pie chart components
* <b>wedges:</b> A list of Patch objects representing each wedge
* <b>texts:</b> List of Text objects representing the labels
* <b>autotexts:</b> List of Text objects for the numeric values - this is only availabe if the autopct value for the pie chart is not None

```python
wedges,texts,autotexts=plt.pie(df_pie['Percentage'],
       labels=df_pie['Sector'],
                              autopct='%.1f')
plt.title('Pie Chart of Movies Dataset')
plt.axis('equal')
print('wedges',wedges)
print('texts',texts)
print('autotexts',autotexts)
plt.show()
```

#### Pie chart customizations
 We set the values for the colors and autopct properties. The latter sets the format for the values to be displayed. 

https://matplotlib.org/stable/gallery/color/named_colors.html

```python
colors_new=['darkslategrey','teal','darkturquoise','paleturquoise','lightcyan']
plt.pie(df_pie['Percentage'],
       labels=df_pie['Sector'],
       colors=colors_new)
plt.title('Pie Chart of Movies Dataset')
plt.axis('equal')
plt.show()
```

#### The explode property
To highlight a particular wedge of the pie chart, we use explode to separate it from the rest of the chart. 

The value for "explode" represents the fraction of the radius with which to offset each wedge.

```python
explode_pie=[0,.1,.2,0,.3]
colors_new=['darkslategrey','teal','darkturquoise','paleturquoise','lightcyan']
plt.pie(df_pie['Percentage'],
       labels=df_pie['Sector'],
       colors=colors_new,
       explode=explode_pie)
plt.title('Pie Chart of Movies Dataset')
plt.axis('equal')
plt.show()
```

## 3 Time Series by line plot
* Time series is a line plot and it is basically connecting data points with a straight line. 
* It is useful in understanding the trend over time. 
* It can explain the correlation between points by the trend. 
* An upward trend means positive correlation and downward trend means a negative correlation. 
* It mostly used in forecasting, monitoring models.

### Compare stocks side by side

```python
fig=plt.figure(figsize=(10,6))
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([.05,.65,.5,.3])
ax1.set_title('AAPL vs IBM (inset)')
ax1.plot(df_line['Date'],
        df_line['AAPL'],
        color='green')
ax2.plot(df_line['Date'],
        df_line['IBM'],
        color='blue')

plt.show()
```

### Compare multiple stocks

## 4 Boxplot 
Boxplot gives a nice summary of the data. It helps in understanding our distribution better.

* The end of the box shows the upper quartile(75%)and the start of the box represents the lower quartile(25%). 
* Upper Quartile is also called 3rd quartile and similarly, Lower Quartile is also called as 1st quartile. 
* The region between lower quartile and the upper quartile is called as Inter Quartile Range(IQR) and it is used to approximate the 50% spread in the middle data(75â€“25=50%).
